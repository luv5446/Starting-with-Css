```python
 from Ipython.display Import HTML

1. Create a simple page with some div tags and show different ways to add CSS as well as what happens
when you target the same elements with inline, internal, and external CSS. Also, utilize comments in the
project where required. 
 
```

```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>CSS Priority Demo</title>

    <!-- Link to external CSS -->
    <link rel="stylesheet" href="styles.css">

    <style>
        /* Internal CSS */
        .box {
            background-color: lightblue; /* This will be overridden by inline CSS if used */
            color: darkblue;
            padding: 20px;
        }
        /* External CSS */

.box {
    background-color: lightgreen; /* overridden by internal or inline */
    color: black;
    border: 2px solid green;
}

.external-only {
    background-color: #eee;
    color: #333;
    border: 2px dashed gray;
    padding: 20px;
}

    </style>
</head>
<body>

    <!-- This div has inline CSS -->
    <div class="box" style="background-color: orange; color: white;">
        Inline CSS Overrides Internal & External
    </div>

    <!-- This div relies on internal and external styles -->
    <div class="box">
        Internal CSS Overrides External
    </div>

    <!-- This div is affected only by external CSS -->
    <div class="external-only">
        External CSS Only
    </div>

    <!-- Comments are helpful to explain code sections -->
    <!-- End of content boxes -->

</body>
</html>
```

```python
2. Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using
class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and
explain how you've named the classes.
```

```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <title>BEM Paragraph Styling</title>
  
  <style>
  /* Base block style */
.paragraph {
  font-family: Arial, sans-serif;
  font-size: 16px;
  line-height: 1.6;
  margin-bottom: 20px;
}

/* Modifier: --intro */
.paragraph--intro {
  font-size: 20px;
  color: #2a5298;
}

/* Modifier: --highlight */
.paragraph--highlight {
  background-color: yellow;
  color: #333;
  padding: 10px;
}

/* Modifier: --note */
.paragraph--note {
  font-style: italic;
  border-left: 4px solid #888;
  padding-left: 12px;
  color: #555;
}

/* Modifier: --error */
.paragraph--error {
  color: red;
  font-weight: bold;
  background-color: #ffe6e6;
  border: 1px solid red;
  padding: 10px;
}

  </style>
</head>
<body>

  <h1>Styled Paragraphs with BEM</h1>

  <!-- Block: paragraph -->
  <!-- Modifier: --intro -->
  <p class="paragraph paragraph--intro">
    This is an intro paragraph with a larger font and different color.
  </p>

  <!-- Modifier: --highlight -->
  <p class="paragraph paragraph--highlight">
    This paragraph is highlighted with a yellow background and dark text.
  </p>

  <!-- Modifier: --note -->
  <p class="paragraph paragraph--note">
    This is a note-style paragraph with italic text and a border.
  </p>

  <!-- Modifier: --error -->
  <p class="paragraph paragraph--error">
    This is an error message styled in red with bold text.
  </p>

</body>
</html>


```

```python
3.  Develop an HTML form with various input elements. Use CSS to style the form, including setting
background colors for input fields. Create a custom color palette for the form elements, and demonstrate
how to apply opacity to one of the form sections.
```

```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <title>Styled HTML Form</title>
 
  <style>
 
:root {
  --color-bg: #f4f7f8;
  --color-primary: #005f73;
  --color-accent: #0a9396;
  --color-input-bg: #e0fbfc;
  --color-border: #94d2bd;
  --color-text: #001219;
}

body {
  font-family: 'Segoe UI', sans-serif;
  background-color: var(--color-bg);
  color: var(--color-text);
  padding: 40px;
}

h1 {
  color: var(--color-primary);
}


.form {
  max-width: 500px;
  margin: auto;
  background: white;
  padding: 30px;
  border-radius: 8px;
  box-shadow: 0 0 10px rgba(0, 0, 0, 0.1);
}


.form__section {
  margin-bottom: 20px;
  display: flex;
  flex-direction: column;
}


.form__section--faded {
  opacity: 0.6; /* translucent effect */
}


label {
  margin-bottom: 8px;
  font-weight: 500;
}


input[type="text"],
input[type="email"],
input[type="password"],
select,
textarea {
  padding: 10px;
  border: 2px solid var(--color-border);
  border-radius: 4px;
  background-color: var(--color-input-bg);
  font-size: 1rem;
}


button {
  background-color: var(--color-primary);
  color: white;
  padding: 10px 20px;
  font-size: 1rem;
  border: none;
  border-radius: 4px;
  cursor: pointer;
}

button:hover {
  background-color: var(--color-accent);
}

  </style>
</head>
<body>

  <h1>Registration Form</h1>

  <form class="form">

    <!-- Section with opacity -->
    <div class="form__section form__section--faded">
      <label for="name">Full Name:</label>
      <input type="text" id="name" name="name" placeholder="Enter your name">
    </div>

    <div class="form__section">
      <label for="email">Email:</label>
      <input type="email" id="email" name="email" placeholder="Enter your email">
    </div>

    <div class="form__section">
      <label for="password">Password:</label>
      <input type="password" id="password" name="password" placeholder="Enter your password">
    </div>

    <div class="form__section">
      <label for="role">Role:</label>
      <select id="role" name="role">
        <option value="user">User</option>
        <option value="admin">Admin</option>
      </select>
    </div>

    <div class="form__section">
      <label><input type="checkbox" name="subscribe"> Subscribe to newsletter</label>
    </div>

    <div class="form__section">
      <label for="comments">Comments:</label>
      <textarea id="comments" name="comments" rows="4" placeholder="Write your message..."></textarea>
    </div>

    <div class="form__section">
      <button type="submit">Submit</button>
    </div>

  </form>

</body>
</html>
```
